```python
import ultralytics
from ultralytics import YOLO


model = YOLO("yolo26s.pt")
```

```python
# Train
model.train(
    data="data.yaml",
    epochs=100,
    patience = 10,
    imgsz=640,
    batch=16, 
    workers=8,
    task="detect",
    optimizer="MuSGD",
    amp=True,
    dropout=0.1,
    freeze=10,
    lr0=0.01,
    lrf=0.01,
    #cls=0.5,
    #box=10.0,
    #dfl=1.5,
    hsv_h=0.015, hsv_s=0.4, hsv_v=0.4,
    degrees=5, translate=0.1, scale=0.3,
    mosaic = 0.0, close_mosaic=10, shear = 0.0, perspective = 0.0, erasing=0.1,
    fliplr=0.5, flipud=0.0,
    mixup=0.0, copy_paste=0.0,
    warmup_epochs=3.0,
    save_period=100,
    conf=0.25,
    device = "cuda:0",
    pretrained = True,
    verbose = True,
    val=True,
    cos_lr=True,
    cache = False,
    plots=True,
)
```

```python
# Tuning
model.tune(
    data='data.yaml', 
    epochs=30, 
    iterations=50, 
    optimizer='MuSGD', 
    plots=True, 
    save_period=30,
    save=True, 
    val=False
)
```
